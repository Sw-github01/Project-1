```python
# Assess of numerical data selected
# Percentage of values missing greater than 75%?
most_missing_cols=df['FormalEducation'].isnull().mean() > 0.75
most_missing_cols
```

```python
# Assess of numerical data selected
# Proportion of individuals in the dataset with career satisfaction reported
prop_sals = 1 - df.isnull()['FormalEducation'].mean()

prop_sals
```

```python
# Assess of numerical data selected
# Count number of row with each unique value of variable
careerSat_vals = df.CareerSatisfaction.value_counts()
careerSat_vals
```

```python
# Assess of numerical data selected
# Value distribution of the numerical data 
# Summary statistic of the numerical data
df['CareerSatisfaction'].describe().reset_index()
```

```python
percent_FormEd=df.FormalEducation.value_counts()/df.FormalEducation.shape[0]-df.FormalEducation.isnull().sum()
percent_FormEd
```

```python
# Using statistics to answer questions and draw insight --- grouping/ungrouping feature?
higher_ed = lambda x: 'Yes' if x in ("Master's degree", "Doctoral", "Professional degree","Bachelor's degree") else 'No'
df['HigherEd'] = df["FormalEducation"].apply(higher_ed)
df[['HigherEd','FormalEducation']].describe().reset_index()
```
